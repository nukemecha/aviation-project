```python
import pandas as pd
```

```python
csv_path ="cleaned_final.csv"

# Import the books.csv file as a DataFrame
aviation_df = pd.read_csv(csv_path, encoding="utf-8",parse_dates=["Event_Date"])
aviation_df.head()
```

```python
aviation_df["Year"]=aviation_df["Event_Date"].dt.year
aviation_df
```

```python
final_df=aviation_df.groupby(["Year","Country"]).sum().reset_index()
final_df
```
