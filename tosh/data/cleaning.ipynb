```python
import pandas as pd
```

```python
df = pd.read_csv('AviationData.csv', encoding = "ISO-8859-1")
```

```python
df.head()
```

```python
cols_to_drop = ['Investigation.Type','Accident.Number','Airport.Name','Registration.Number','Amateur.Built',\
               'Number.of.Engines', 'Engine.Type', 'FAR.Description', 'Schedule', 'Purpose.of.Flight',\
               'Air.Carrier', 'Weather.Condition','Report.Status','Publication.Date']
```

```python
df.drop(cols_to_drop, 1, inplace=True)
```

```python
df.head()
```

```python
# fill NaN in injury columns with zeroes as they should be. 

df['Total.Fatal.Injuries'].fillna(0, inplace=True)
```

```python
df['Total.Serious.Injuries'].fillna(0, inplace=True)
df['Total.Minor.Injuries'].fillna(0, inplace=True)
df['Total.Uninjured'].fillna(0, inplace=True)
```

```python
df.head()
```

```python
# fill NaN in flight phase with UNKNOWNs. 

df['Broad.Phase.of.Flight'].fillna('UNKNOWN', inplace=True)
```

```python
df.head()
```

```python
# drop accidents with no coordinates

df.dropna(subset=['Latitude'], inplace=True)
```

```python
df.head()
```

```python
# fill NaN Airport codes

df['Airport.Code'].fillna('UNKNOWN', inplace=True)
```

```python
# fill NaN Aircraft categories

df['Aircraft.Category'].fillna('Unknown', inplace=True)
```

```python
df['Make'].fillna('Unknown', inplace=True)
df['Model'].fillna('Unknown', inplace=True)
```

```python
df['Location'].fillna('Unknown', inplace=True)
df['Country'].fillna('Unknown', inplace=True)
df['Injury.Severity'].fillna('Unknown', inplace=True)
df['Aircraft.Damage'].fillna('Unknown', inplace=True)
```

```python
df.sample(10)
```

```python
df.to_csv('cleaned_data.csv')
```
