```python
import pandas as pd
```

```python
df = pd.read_csv('cleaned_data2.csv')
```

```python
df.drop(['Unnamed: 0'], 1, inplace=True)
```

```python
df.head()
```

```python
df['Event.Date'] = pd.to_datetime(df['Event.Date'])
```

```python
df.head()
```

```python
start_date = '01-01-1982'
end_date = '04-08-2020'
```

```python
mask = (df['Event.Date'] >= start_date) & (df['Event.Date'] <= end_date)
```

```python
ndf = df.loc[mask]
```
